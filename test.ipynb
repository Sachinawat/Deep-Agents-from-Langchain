```python
a=[1,2,1,3,4,5,6,23,4,2,4,1,4,2,4,2,4,55,3]

def remove_duplicates(input_list):
    seen=set()
    output=[]
    for item in input_list:
        if item not in seen:
            seen.add(item)
            output.append(item)
    return output
b=remove_duplicates(a)
print(b)
```
